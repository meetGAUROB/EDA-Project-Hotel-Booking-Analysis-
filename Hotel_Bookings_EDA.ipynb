````python
# Write your code to make your dataset analysis ready.

# Removing outliers.
hotel_df = hotel_df[hotel_df['adr'] <= hotel_df['adr'].quantile(0.90)].reset_index(drop=True)

#Removing duplicates.
hotel_df = hotel_df.drop_duplicates(keep='first').reset_index(drop=True)

# Convert columns with intger 1/0 into object.
hotel_df['is_canceled'] = hotel_df['is_canceled'].apply(lambda x: 'Canceled' if x==1 else 'Not-Canceled')
hotel_df['is_repeated_guest'] = hotel_df['is_repeated_guest'].apply(lambda x: 'Yes' if x==1 else 'No')

#Making a new column name 'total nights' by adding weekend nights and week nights.
hotel_df['Total Nights'] = hotel_df['stays_in_weekend_nights'] + hotel_df['stays_in_week_nights']

#Making a new column name 'Price' to know the total gain from each booking.
hotel_df['Booking Income'] = hotel_df['adr'].fillna(0) * hotel_df['Total Nights']

# Filling NaN with 0 in children column.
hotel_df['children'] = hotel_df['children'].fillna(0)

#Converting columns which has 0 in it into NaN.
into_nan = ['lead_time', 'stays_in_weekend_nights',
            'stays_in_week_nights', 'adults', 'children',
            'babies', 'previous_cancellations', 'previous_bookings_not_canceled',
            'booking_changes', 'days_in_waiting_list', 'adr','Booking Income', 'Total Nights', 'required_car_parking_spaces', 'total_of_special_requests']
hotel_df[into_nan] = hotel_df[into_nan].replace(0, np.nan)
try:
  # Converting reservation status date into datetime format.
  hotel_df['reservation_status_date'] = pd.to_datetime(hotel_df['reservation_status_date'])
except Exception as e:
    print("An error occurred while converting reservation status date to datetime format:", e)
````

#### Chart - 20--> Special Request Frequencies by Night Type(Barplot)

````python
#Data Preparation
#filtering out canceled bookings
booking_filter = hotel_df[hotel_df['is_canceled']=='Not-Canceled'].reset_index(drop=True)

#Making two df for week stay and weekend stay.
week_stays = booking_filter[['stays_in_week_nights', 'total_of_special_requests']].reset_index()
weekend_stays = booking_filter[['stays_in_weekend_nights', 'total_of_special_requests']].reset_index()

#filtering out the NaN values from both df.
filtering_week_nan = week_stays.dropna(subset=['stays_in_week_nights']).reset_index(drop=True)
filtering_weekend_nan = weekend_stays.dropna(subset=['stays_in_weekend_nights']).reset_index(drop=True)

#Grouping by total of special requests by counting their ocurances.
group_Week_req = filtering_week_nan.groupby('total_of_special_requests')['stays_in_week_nights'].count().reset_index()
group_Weekend_req = filtering_weekend_nan.groupby('total_of_special_requests')['stays_in_weekend_nights'].count().reset_index()

# merged the two df and rename the column.
merged_request = pd.merge(group_Week_req, group_Weekend_req, on='total_of_special_requests', how='inner')
merged_request.columns = ['Number_of_special_requests', 'Week_nights', 'Weekend_nights']

# melt the df to make the week and weekend columns into values of night type.
melt_df = merged_request.melt(id_vars='Number_of_special_requests', var_name='Night_type', value_name='Ocurances')

# #Chart 20 Visualization code.
plt.figure(figsize=(12,6))
sns.barplot(data=melt_df, x='Number_of_special_requests', y='Ocurances', hue='Night_type')
plt.title('Special Request Frequencies by Night Type')
plt.xlabel('Number of special requests')
plt.ylabel('Frequencies of special requests')
plt.show()
````
